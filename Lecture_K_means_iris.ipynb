```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(df["sepal length (cm)"],df["sepal width (cm)"])
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

```python
KM=KMeans(n_clusters=3)
```

```python
SL=np.array(df["sepal length (cm)"])
SW=np.array(df["sepal width (cm)"])
PL=np.array(df["petal length (cm)"])
PW=np.array(df["petal width (cm)"])
```

```python
input=np.column_stack((SL,SW,PL,PW))
```

```python
KM.fit(input)
```

```python
KM.labels_
```

```python
plt.scatter(df["sepal length (cm)"],df["sepal width (cm)"],c=KM.labels_)
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

```python
plt.scatter(df["petal length (cm)"],df["petal width (cm)"],c=KM.labels_)
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.show()
```

```python
fig=plt.figure()
axis=fig.add_subplot(projection='3d')

axis.scatter(df["sepal length (cm)"],df["sepal width (cm)"],df["petal length (cm)"],c=KM.labels_)
axis.set_xlabel("sepal length (cm)")
axis.set_ylabel("sepal width (cm)")
axis.set_zlabel("petal length (cm)")
plt.show()
```

```python
df['Kmeans_Result']=KM.labels_
df.head()
```
