```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
cancer=datasets.load_breast_cancer()
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df['target']=cancer.target
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
cancer_scaled=SS.fit_transform(cancer.data)
input=cancer_scaled
output=df['target']
```

```python
from sklearn.model_selection import train_test_split
train_input,test_input,train_output,test_output=train_test_split(input,output)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
DNN=Sequential()
DNN.add(Dense(100,activation="relu"))
DNN.add(Dense(1,activation="sigmoid"))
DNN.compile(loss="mean_squared_error",metrics='accuracy',optimizer="adam")#RMSprop
```

```python
DNN.fit(train_input,train_output,epochs=10, batch_size=50)
```
